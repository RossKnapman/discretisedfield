# Tutorial 3: Mesh plotting

There are two ways how a mesh can be plotted. Using

- [`matplotlib`](https://matplotlib.org/) (`discretisedfield.Mesh.mpl`) or
- [`k3d`](https://github.com/K3D-tools/K3D-jupyter) (`discretisedfield.Mesh.k3d`)

For plotting inside Jupyter, the inline plotting (`%matplotlib inline`) should be enabled after `discretisedfield` is imported.

```python
import discretisedfield as df
%matplotlib inline
```

Firstly, let us create a mesh which we are going to plot using `matplotlib` and `k3d`

```python
p1 = (-6, -3, -3)
p2 = (6, 3, 3)
n = (12, 6, 6)
mesh = df.Mesh(p1=p1, p2=p2, n=n)
```

A `matplotlib` representation of the mesh is

```python
mesh.mpl()
```

Blue solid lines are the edges of the mesh domain, whereas the dashed red lines show the relative size of a single discretisation cell. To change the size of the plot, `figsize` parameter can be passed. It is a tuple of size 2.

```python
mesh.mpl(figsize=(10, 6))
```

In order to get an interactive 3D plot inside Jupyter notebook, `discretisedfield.Mesh.k3d` method should be used.

```python
mesh.k3d()
```

In this interactive 3D plot the mesh domain is represented in blue, whereas a single discretisation cell is represented in red. Different colours can be passed via `colormap` list of length 2, where the first element is the colour of the domain and the second of the discretisation cell. Similarly, outlines can be activated by passing `outlines=True`. For instance

```python
mesh.k3d(colormap=[0x95a5a6, 0xf1c40f])
```

Other parameters that can be passed to this method can be found in [`k3d.voxels documentation`](https://k3d-jupyter.readthedocs.io/en/latest/k3d.html#k3d.k3d.voxels).
