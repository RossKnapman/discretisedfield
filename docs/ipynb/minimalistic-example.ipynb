# Minimalistic plotting example

```python
import discretisedfield as df
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# NBVAL_IGNORE_OUTPUT
mesh = df.Mesh(p1=(-5, -5, -5), p2=(6, 6, 6), cell=(1, 1, 1))
mesh
```

```python
# NBVAL_IGNORE_OUTPUT
def val(pos):
    x, y, z = pos
    return y, x, x*y

field = df.Field(mesh, value=val)
```

## Plotting scalar component (x, y, z)

```python
# NBVAL_IGNORE_OUTPUT
field.x.plot_plane("z")
```

```python
# NBVAL_IGNORE_OUTPUT
field.y.plot_plane("z")
```

```python
# NBVAL_IGNORE_OUTPUT
field.z.plot_plane("z")
```

## Plotting vector

```python
# NBVAL_IGNORE_OUTPUT
field.plot_plane("z")
```

## Divided plotting

```python
# NBVAL_IGNORE_OUTPUT
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

imax = field.z.imshow("z", ax=ax, interpolation="nearest", cmap="viridis")
qvax = field.quiver("z", ax=ax, n=(11, 11))
field.colorbar(ax, imax)

ax.set_title("Plot slice")
ax.set_xlabel("x label")

ax.set_ylabel("y label")
```
