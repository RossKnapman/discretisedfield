```python
import discretisedfield as df
import matplotlib.pyplot as plt
%matplotlib inline

mesh = df.Mesh(p1=(-1,-1,-1), p2=(2,1,1), cell=(0.2,0.2,0.2))

def val(pos):
    x, y, z = pos
    #if x == 0.5:
    #    return (100, 0, 0)
    return y, x, x*y
    

field = df.Field(mesh, value=val)
```

```python
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(111)

ax = field.plot_slice(y=0, ax=ax);

ax.set_xlabel("AAA");
```
