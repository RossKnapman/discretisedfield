There are two ways how a mesh can be plotted. Using

- matplotlib (`discretisedfield.Mesh.mpl`) or
- k3d (`discretisedfield.Mesh.k3d`)

For plotting inside Jupyter, the inline plotting (`%matplotlib inline`) should be enabled after `discretisedfield` is imported.

```python
import discretisedfield as df
%matplotlib inline
```

Now, let us create a mesh which we are going to plot using `matplotlib` and `k3d`

```python
p1 = (-6, -3, -1)
p2 = (6, 3, 1)
n = (12, 6, 2)
mesh = df.Mesh(p1=p1, p2=p2, n=n)
```

A `matplotlib` representation of the mesh is

```python
# NBVAL_IGNORE_OUTPUT
mesh.mpl()
```

Blue solid box is the mesh domain, whereas the dashed red box shows the relative size of a single discretisation cell. To change the size of the plot, `figsize` parameter can be passed. It is a tuple of size 2.

```python
# NBVAL_IGNORE_OUTPUT
mesh.mpl(figsize=(10, 6))
```

In order to get an interactive 3D plot inside Jupyter notebook, `discretisedfield.Mesh.k3d` method should be used.

```python
# NBVAL_IGNORE_OUTPUT
mesh.k3d()
```

In this interactive 3D plot the mesh domain is represented in blue, whereas a single discretisation cell is represented in red. Different colours can be passed via `colormap` list of length 2, where the first element is the colour of the domain and the second of the discretisation cell. Colours are defined in a hexadecimal format. Similarly, outlines can be activated by passing `outlines=True`. Other parameters that can be passed to this method can be found in `k3d.voxels` [documentation`](https://k3d-jupyter.readthedocs.io/en/latest/k3d.html#k3d.k3d.voxels).

```python
# NBVAL_IGNORE_OUTPUT
mesh.k3d(colormap=[0xecf0f1, 0xf1c40f])
```

Similarly, we can plot the points at centres of all discretisation cells in the mesh.

```python
# NBVAL_IGNORE_OUTPUT
mesh.k3d_points()
```

What other parameters can be passed to this function can be found in `k3d.points` [documentation](https://k3d-jupyter.readthedocs.io/en/latest/k3d.html#k3d.k3d.points).

To visualise the defined regions in a mesh, `discretisedfield.Mesh.k3d_regions` should be used. There are 6 predefined colours in the `colormap` argument. If the mesh contains more than 6 regions, all subsequent colors will be generated randomly. Therefore, in the following example, the colours for regions `r7` and `r8` will be generated randomly.

```python
# NBVAL_IGNORE_OUTPUT
p1 = (0, 0, 0)
p2 = (100, 80, 10)
cell = (2, 2, 2)
regions = {'r1': df.Region(p1=(0, 0, 0), p2=(100, 10, 10)),
           'r2': df.Region(p1=(0, 10, 0), p2=(100, 20, 10)),
           'r3': df.Region(p1=(0, 20, 0), p2=(100, 30, 10)),
           'r4': df.Region(p1=(0, 30, 0), p2=(100, 40, 10)),
           'r5': df.Region(p1=(0, 40, 0), p2=(100, 50, 10)),
           'r6': df.Region(p1=(0, 50, 0), p2=(100, 60, 10)),
           'r7': df.Region(p1=(0, 60, 0), p2=(100, 70, 10)),
           'r8': df.Region(p1=(0, 70, 0), p2=(100, 80, 10))}
mesh = df.Mesh(p1=p1, p2=p2, cell=cell, regions=regions)
mesh.k3d_regions()
```

Other parameters that can be passed to this method can be found in `k3d.voxels` [documentation](https://k3d-jupyter.readthedocs.io/en/latest/k3d.html#k3d.k3d.voxels).
