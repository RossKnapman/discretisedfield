# Tutorial 12: `ovf2vtk` tool

> Interactive online tutorial:
> [![Binder](https://mybinder.org/badge_logo.svg)](https://mybinder.org/v2/gh/ubermag/discretisedfield/master?filepath=docs%2Fipynb%2Findex.ipynb)

`discretisedfield.ovf2vtk` is a command line tool which can be used to convert OVF files to the VTK format.

Information about the tool and its usage can be accessed as:

```
$ python3 -m discretisedfield.ovf2vtk --help
usage: ovf2vtk [-h] --infile INFILE [INFILE ...]
               [--outfile OUTFILE [OUTFILE ...]]

ovf2vtk - ovf to VTK format conversion

optional arguments:
  -h, --help            show this help message and exit
  --infile INFILE [INFILE ...]
                        One or more input files
  --outfile OUTFILE [OUTFILE ...]
                        One or more output files, optional
```

There are three main use cases now `ovf2vtk` can be used. In any case, specifying the output file name is optional. More precisely, if the names of the output files are not provided, they will be the same as the input files, but with a different extension (`.vtk` instead of `.ovf`).

### 1. Single file coversion

Probably the simplest use case is the conversion of a single fie.

```
$ python3 -m discretisedfield.ovf2vtk --infile m0.omf
```

This command is going to create the same output file as the following command (when we specify the output file name):

```
$ python ovf2vtk.py --infile m0.omf --outfile m0.vtk
```

### 2. List of files

To convert multiple files using a single command, multiple `.ovf` files can be passed to the `ovf2vtk` tool.

```
$ python3 -m discretisedfield.ovf2vtk --infile m0.omf m1.omf m2.omf
```

### 3. Wildcard

Usually, the number of files which should be converted is very large. In that case, a filename with a wildcard can be passed.

```
$ python3 -m discretisedfield.ovf2vtk --infile m*.omf
```
